# Exploratory Data Analysis (EDA) - Real Estate Listings

Phân tích dữ liệu từ file `listing_details.csv` để hiểu cấu trúc và chất lượng dữ liệu.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Thiết lập hiển thị
%matplotlib inline
pd.set_option('display.max_columns', None)

# Đường dẫn dữ liệu
file_path = '../data/listing_details.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"Đã tải dữ liệu thành công! Kích thước: {df.shape}")
else:
    print("Không tìm thấy file dữ liệu!")
```

## 1. Kiểm tra giá trị thiếu (Missing Values)

Xác định phần trăm dữ liệu bị thiếu trong mỗi cột.

```python
# Tính toán số lượng và phần trăm thiếu
missing_data = df.isnull().sum().to_frame(name='Missing_Count')
missing_data['Percentage (%)'] = (missing_data['Missing_Count'] / len(df)) * 100
missing_data = missing_data.sort_values(by='Percentage (%)', ascending=False)

display(missing_data)
```

### Trực quan hóa giá trị thiếu

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_data.index, y=missing_data['Percentage (%)'], palette='viridis')
plt.xticks(rotation=90)
plt.title('Phần trăm giá trị thiếu theo từng cột')
plt.ylabel('Phần trăm (%)')
plt.show()
```
